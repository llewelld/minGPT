```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('../data/results.csv')
df.head()
```

```python
ax = df.plot.line(x='Embedding size (tokens)', y='Epoch time Baskerville 1 GPU (seconds)', legend=False)
ax.set_title('Epoch time Baskerville 1 GPU')
ax.set_ylabel('Time (seconds)')
plt.show()
```

```python
ax = df.plot.line(x='Embedding size (tokens)', y='Peak memory Baskerville 1 GPU (MiB)', legend=False)
ax.set_title('Peak memory Baskerville 1 GPU')
ax.set_ylabel("Memory (MiB)")
plt.show()
```
